```python
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot
import numpy as np
```

```python
df = pd.read_csv('auto-mpg.csv')
df.drop('origin',axis=1,inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df['acceleration'].fillna((df['acceleration'].mode()[0]),inplace=True)
df['model year'].fillna((df['model year'].mode()[0]),inplace=True)
df['CAR Number'].fillna((df['CAR Number'].mode()[0]),inplace=True)
df.isnull().sum()
```

```python
df.drop_duplicates(subset=['CAR Number'],inplace=True)
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
df.info()
```

```python
df.loc[df['horsepower'] == df['horsepower'].max(),'car name'].to_list()
```

```python
#Q5
df['mpg'] = 235/df["mpg"]
df.rename(columns={'mpg':'L/100km'},inplace=True)
df.head()
```

```python
#Q6
df['weight'] = df['weight']/df['weight'].max()
df.head()
```

```python
#Q10
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
group_names = ['Little horsepower', 'Medium horsepower', 'High horsepower']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
#Q11
hist_bar = df["horsepower-binned"].value_counts()
pyplot.bar(group_names,hist_bar)

plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower binned')
```

```python
#Q12 Outliner may occur in acceleration column
from scipy.stats import zscore
df['zscore'] = zscore(df[['acceleration']])
newdf = df.drop(df.loc[df['zscore'].abs()>3].index)
newdf
```
